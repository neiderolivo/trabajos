para la solucion del problema se utlizaran las siguientes ecuaciones: 

fuerza gravitacional
\begin{equation}
F=\frac{G*M_t*m}{R²} (1)
\end{equation}
fuerza centripeta.

\begin{equation}
 F_c=\frac{mv²}{R}  (2)
\end{equation}
velocidad con la que gira el satelite

\begin{equation}
V=\frac{2\pi R}{T}  (3)
\end{equation}

igualando la  ecuacion (1) y (2) para asi despejar  R, obtenemos lo siguiente:

\begin{equation}
\frac{G*M_t*m}{R²}=\frac{mv²}{R}
\end{equation}

podemos apreciar que se puede simplificar  la masa m y un radio R, por lo tanto:

\begin{equation}
\frac{G*M_t}{R}=v²
\end{equation}

sacando la raiz cuadrada a la velocidad:

\begin{equation}
v=\sqrt{\frac{G*M_t}{R}} (4)
\end{equation}

se obtiene una nueva ecuacion para la velocidad llamada 4 ahora dicha ecuacion la igualamos con la ecuacion 3,obteniendo:

\begin{equation}
\frac{2\pi R}{T}=\sqrt{\frac{G*M_t}{R}} 
\end{equation}

utlizando algebra para despejar el valor de R se obtiene que dicho valor se puede calcular con la siguiente exprexion.

\begin{equation}
R=\sqrt[3]{\frac{G*M_t*T²}{4\pi²}}
\end{equation}

dicho radio equivale a la distancia desde el centro de la tierra al satelite, pero como se desea calcular solo desde la superficie de la tierra al satelite, se debe utilizar la siguiente expresion.

\begin{equation}
H=R-R_t\\
R_t= radio de la tierra
\end{equation}
